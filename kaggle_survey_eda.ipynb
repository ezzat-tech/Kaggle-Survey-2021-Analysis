# Kaggle 2021 ML & Data Science Survey EDA Project

## Objective: 

### To identify the most widely used tools and platforms in the machine learning and data science community and comparing usage between students and industry professionals.

## 1. Loading the Data

```python
import pandas as pd

ks = pd.read_csv('kaggle_survey_2021_responses.csv')
pd.set_option('display.max_columns', None)
```

## 2. EDA & Data Cleaning

**As per the survey methodology, the dataset is free from duplicates or spam responses which provides us with a clean and high quality data to deal with.**

*Since the Dataset provided includes 25,973 rows and 369 columns, including responses with multiple choices can be selected split into a dozen columns, only selected survey questions columns will be used to achieve the objective mentioned.*



```python
ks
```

```python
# Renamed the series into interpretable names for easier understanding and usage throughout the analysis that has only one choice selected i.e one column per question.

ks.rename(columns={
    'Q1': 'age_category',
    'Q2': 'gender',
    'Q3': 'country',
    'Q4': 'education_level',
    'Q5': 'current_role',
    'Q6': 'years_programming',
    'Q8': 'recommended_first_language',
    'Q11': 'primary_computing_platform',
    'Q13': 'tpu_usage_frequency',
    'Q15': 'years_using_ml',
    'Q20': 'industry',
    'Q21': 'company_size',
    'Q22': 'num_data_science_staff',
    'Q23': 'ml_adoption_status',
    'Q25': 'yearly_compensation_usd',
    'Q26': 'ml_cloud_spending_5y_usd',
    'Q28': 'most_enjoyable_cloud_platform',
    'Q33': 'most_used_big_data_product',
    'Q35': 'most_used_business_intelligence_tool',
    'Q41': 'primary_data_analysis_tool'

                  }
          , inplace=True)
```

```python
ks.drop(index=0)
ks
```

```python
# Created a new dataframe with the data that has only one choice selected i.e one column per question and asked to both groups.


main_data = ks[[
    'age_category',
    'gender',
    'country',
    'education_level',
    'years_programming',
    'current_role',
    'recommended_first_language',
    'primary_computing_platform',
    'tpu_usage_frequency',
    'years_using_ml',
    'industry',
    'company_size',
    'num_data_science_staff',
    'ml_adoption_status',
    'yearly_compensation_usd',
    'ml_cloud_spending_5y_usd',
    'primary_data_analysis_tool'
]]
main_data
```

```python
# Created a function to categorize reviewers into two main categories: Students or Professionals

def student_or_not(current_role):
    if current_role == 'Student':
        return 'Students'
    else:
        return 'Professionals'
    
```

```python
main_data['role_category'] = main_data['current_role'].apply(student_or_not)
main_data
```

## 3. Analysis

```python
main_data.groupby('role_category')[['age_category','education_level','industry','ml_adoption_status','years_using_ml','primary_computing_platform']].agg(pd.Series.mode)
```

*- Most students answering the survey fall in the 18-21 age category and industry professionals in the 25-29 age category*

*- Both groups utilize a laptop as the main device

*- Both groups have very little experience in ML but industry professionals are exploring ML methods to adopt a model into production*

```python
# Created a seperate dataframe with Students only including a filter that drops the questions they were not asked or were asked an alternative question, for example, students were asked 27-B instead of 27_A.

Students_df = ks[ks['current_role']=='Student'].filter(regex='^(?!.*_A)').drop(columns=['most_enjoyable_cloud_platform', 'most_used_big_data_product', 'most_used_business_intelligence_tool'])
Students_df
```

```python
# Created a seperate dataframe with Industry Professionals only including a filter that drops the questions they were not asked or were asked an alternative question, for example, Industry Professionals were asked 27-A instead of 27-B.
Professionals_df = ks[ks['current_role']!='Student'].filter(regex='^(?!.*_B)')
Professionals_df
```

```python
# Professionals across all age categories recommended python as the first language to learn to get into Data Science & ML
Professionals_df.groupby('age_category')[['recommended_first_language']].agg(pd.Series.mode)
```

### 3.1 Tools usage Comparison Analysis between students & Industry Professionals

```python
# Understanding what tools professionals in the industry mostly use & enjoy

Professionals_df.groupby('age_category')[['most_enjoyable_cloud_platform','most_used_big_data_product','most_used_business_intelligence_tool']].agg(pd.Series.mode)
```

```python
# Created a function to use it to get the percentages of answers of each question with multiple choices selected for the comparison between students & professionals
def get_pct_multiple(df, question,labels,group_name):
    result = pd.DataFrame({group_name: (df.filter(like=question).notna().sum()/len(df)*100)}).round(2)
    result.index = labels
    return result
```

```python
# Q7 - What programming languages do you use on a regular basis? (Select all that apply)

language_labels = ['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'None', 'Other']

q7_students = get_pct_multiple(Students_df, 'Q7', language_labels, 'Students')
q7_professionals = get_pct_multiple(Professionals_df, 'Q7', language_labels, 'Professionals')

# Concatenated both results into one Dataframe for easier comparison

q7_results = pd.concat([q7_students,q7_professionals], axis=1)

q7_results.head(5)#.plot.barh(xlabel='% of selection')
```
